### Now we need to know where are the areas that we need to work with to make/improve profits.

**Q1: What are the states that we should start working on thier profits numbers?**

```python
profit_mean = df.Profit.mean()
weak_states = df.loc[df['Profit'] <= profit_mean].groupby('State').size().sort_values(ascending=False)
weak_states
```

#### We got the mean of each state profits to deal with the profits that less than the mean.

```python
weak_states.plot(kind='barh', figsize=(12, 10), title="Weak States vs Profits");
```

```python
df.loc[df['State']=='California'].shape[0],df.loc[df['State']=='West Virginia'].shape[0]
```

### This visualization shows how many single profits less than the mean profit each state have.
    - California cities has the highest number of profits less than the mean profit.
    - West Virginia cities has the lowest number of profits less than the mean profit.
### But keep in mind the number of rows for each state as California has 2001 rows of profits while West Virginia has 4 rows of profits.

**Q2: Which region has the highest profits?**

```python
df.Region.unique()
```

```python
regions_by_profits = df.groupby('Region')['Profit'].sum()
```

```python
fig, ax = plt.subplots(figsize=(12,6))
label = ['South', 'West', 'Central', 'East']
ax.pie(regions_by_profits, labels=label, autopct='%.1f%%',shadow=True, textprops={'color':'White', 'fontsize': 18}, radius=1.1)
ax.set_title('Regions by Profits', fontdict={'fontsize':15})
plt.legend(title='Regions', loc='upper right', fontsize=12)
plt.tight_layout()
plt.show();
```

### Pie chart above shows the percentages of profits for each region in the US.
    - The US East region has the highest percentage of profit.

**Q3: What is the most and least sold Category?**

```python
category_by_profit = df.groupby('Category')['Profit'].sum()
category_by_profit
```

```python
fig, ax = plt.subplots(figsize=(12,6))
ax.pie(category_by_profit, labels=category_by_profit.index, autopct='%.1f%%',shadow=True, textprops={'color':'White', 'fontsize': 18}, radius=1.1)
ax.set_title('Category by Profits', fontdict={'fontsize':15})
plt.legend(title='Categories', loc='upper right', fontsize=12)
plt.tight_layout()
plt.show();
```

### The most sold category is Technologies, and the least sold category is Furniture.

**Q4: What is the most and least sold Sub-Category?**

```python
subcategory_by_profit = df.groupby('Sub-Category')['Profit'].sum()
subcategory_by_profit
```

```python
# Create the bar chart
plt.bar(subcategory_by_profit.index, subcategory_by_profit.values, color = ['green' if x>0 else 'red' for x in subcategory_by_profit.values],width=1)
plt.xlabel("Sub-Category")
plt.ylabel("Profit")
plt.title("Profit by Sub-Category", fontdict={'fontsize':15})
plt.grid()
plt.xticks(rotation=90)
plt.show();
```

### There's profits in negative and so low positive profit values.
    - The sales team need to deal with the Tables profits as it's seems like a big loss to the company.
    - Also, the sales team need to take in consideration the Bookcases, Supplies, Fasteners and Machines.
